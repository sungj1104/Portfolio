# 문제3. 문자열 데이터를 입력 받고 capitalize() 메소드를 적용하는 capitalizer 함수 정의

```python
def capitalizer(n):
    return n.capitalize()
```

# 문제4. Mjob과 Fjob열에 capitalizer 함수 적용된 결과 출력

```python
df[["Mjob", "Fjob"]].applymap(capitalizer)
```
