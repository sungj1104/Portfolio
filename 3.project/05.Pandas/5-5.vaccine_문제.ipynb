# 문제4. day1num과 day2num의 월별 평균을 출력하기

```python
data = df.copy()
```

```python
df["y_m"] = df["date"].str[:7]
```

```python
df[["y_m", "day1num", "day2num"]].groupby("y_m").mean()
```

```python
for i in data.index:
    data["date"][i] = data["date"][i][:7]

data.groupby("date").mean()[["day1num", "day2num"]]
```
