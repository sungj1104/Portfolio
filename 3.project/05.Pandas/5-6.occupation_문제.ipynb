```python
import pandas as pd
```

# 문제1. user_id를 인덱스로 사용한 데이터프레임 읽기

```python
df = pd.read_csv("./data/occupation.tsv", sep = "\|", index_col = "user_id")
```

```python
df.head()
```

# 문제2. 각 occupation별 age의 평균 구하기

```python
df.groupby("occupation")["age"].mean()
```

# 문제3. 각 occupation 별 남자의 비율 구하기
- 단, 남자 비율의 내림차순으로 정렬하여 출력하기

```python
item = df.groupby("occupation")["gender"].value_counts(normalize = True)
```

```python
j = dict()
for i in item.index:
    if i[1] == "M":
        j[i[0]] = item[i]
```

```python
data = pd.DataFrame([j]).transpose()[0].sort_values(ascending = False)
```

```python
data = pd.DataFrame(data)
data.columns = ["각 occupation 별 남자의 비율"]
data
```

```python
(df[df["gender"] == "M"].groupby("occupation").count() / df.groupby("occupation").count())["gender"].sort_values(ascending = False)
```

```python
df["male_n"] = df["gender"].map(lambda x: 1 if x == "M" else 0)
(df.groupby("occupation")["male_n"].sum() / df["occupation"].value_counts()).sort_values(ascending = False)
```

# 문제4. occupation별 age의 최솟값, 최댓값 구하기

```python
df.groupby("occupation")["age"].agg(["min", "max"])
```

# 문제5. 각 occupation의 gender별 평균 age를 구하기

```python
df.groupby(["occupation", "gender"])["age"].mean()
```
