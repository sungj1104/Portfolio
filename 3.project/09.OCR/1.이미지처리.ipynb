# 이미지 처리(image processing)

- 아날로그 이미지 처리
    - 사진 촬영 시 필름에 다양한 화학 처리르 하거나 카메라로 촬영한 이미지를 변경 및 편집하기 위해 물리적 필터를 사용하는 등 물리적 수단을 이용해 이미지를 조작하거나 편집하는 것
    
- 디지털 이미지 처리
    - 이미지 처리를 하는데에 컴퓨터와 수학적 알고리즘과 계싼 기술에 의존
    - 포함되는 기술
        - 이미지 향상(image upscaling)
        - 이미지 복원(image restoration)
        - 특징 추출(feature extraction) 등
        
    - 디지털 이미지는 유한한 픽셀 집합으로 구성되며, 각 픽셀은 색상과 강도에 대한 정보를 담고 있음
        - 이미지의 구성 요소인 픽셀(pixel)의 역할을 이해하는 것이 중요
        - 이미지 처리 알고리즘은 픽셀에서 작동하여 이미지에서 중요한 정보를 향상, 변환, 추출할 수 있게 해줌

## 디지털 이미지의 처리 단계

    1. 이미지 획득(image acquisition)
            - 카메라, 스캐너, 또는 다른 이미지 캡처 장치를 통해 이루어짐
            - 획득한 이미지는 디지털 형식으로 변환
                - 이미지의 품질을 최대한 보존하면서 디지털 데이터로 변환하는 것이 중요
                - 이미지 획득 과정에서 발생할 수 있는 노이즈, 왜곡, 빛의 변화 등을 처리하는 기술도 중요

    2. 이미지 개선(image enhancement)
        - 이미지의 품질을 향상시키는 데 초점을 맞춤
        - 노이즈 제거, 명암 조절, 색상 보정 등의 작업을 포함
        - 이미지를 더욱 명확하고 이해하기 쉽게 만드는 것이 목표
            - 이를 위해 다양한 필터링 기법과 히스토그램 평활화, 샤프닝 등의 기술이 사용됨
            - 이미지의 노이즈를 줄이고, 세부사항을 강조하며, 이미지의 전반적인 품질을 향상

    3. 이미지 분석(image analysis)
        - 이미지의 유요한 정보를 추출
        - 특징 추출, 패턴 인식, 객체 감지 등의 작업을 포함
        - 이미지에서 의미있는 데이터를 얻는 것이 목표
            - 이를 위해 에지 검출, 코너 검출, 텍스처 분석 등의 기술이 사용됨
            - 이미지의 구조와 패턴을 파악하고 이미지에서 중요한 특징을 식별하며 이 특징을 이용하여 이미지를 분석

    4. 이미지 해석 및 이해(image interpretation and understanding)
        - 이미지 분석 단계에서 얻은 데이터를 이용하여 이미지를 해석하고 이해
        - 이미지 분류, 이미지 검색, 이미지 인식 등의 작업을 포함
        - 이미지에서 얻은 데이터를 의미있는 방식으로 사용하는 것이 목표
            - 이를 위해 패턴 매칭, 머신러닝, 딥러닝 등의 기술이 사용됨
            - 이미지에서 얻은 데이터를 분석하고 이 데이터를 이용하여 이미지를 분류하거나 특정 패턴을 인식하거나 이미지 내의 객체를 식별하는 등의 작업을 수행

# 컴퓨터 비전

- 기계가 시각적 데이터를 이해하고 분석하는 능력을 개발하는 과학 분야

- 이미지 처리와 컴퓨터 비전의 공통점
    - 디지털 이미지를 사용하여 우리가 세상을 이해하는 방식을 확장하고 변형
    
- 이미지 처리와 컴퓨터 비전의 차이점
    - 이미지 처리
        - 디지털 이미지의 향상, 변형, 복원 등에 중점을 둠
    - 컴퓨터 비전
        - 객체 인식, 패턴 분석, 이미지 분류 등의 작업에 중점
    - 이미지 처리는 이미지의 품질을 개선하고 중요한 정보를 강조하며, 컴퓨터 비전은 정보를 분석하고 해석하여 더 높은 수준의 결론을 도출

## 컴퓨터 비전의 정의

- 시각 기계의 과학 및 기술
    - 이미지에서 정보를 추출하는 인공 시스템 이론
- 이미지 데이터는 비디오 시퀀스, 여러 카메라의 뷰 또는 의료용 스캐너의 다차원 데이터 등 다양한 형태를 취할 수 있음

- 컴퓨터 비전 기술은 낮은 수준, 중간 수준, 높은 수준의 비전 작업으로 분류 할 수 있음
    - 낮은 수준 비전 작업
        - 노이즈 제거
        - 대비 향상
        - 채도 향상
        - 에지 검출
        
    - 중간 수준 비전 작업
        - 이미지 영역 분할
        - 이미지 객체로 분할
        - 이미지 광학 흐름 추정
        
    - 높은 수준 비전 작업
        - 객체 인식
        - 장면 재구성
        - 이미지 학습 및 추론
