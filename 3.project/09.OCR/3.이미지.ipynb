# 디지털 이미지의 구조

- 픽셀들의 집합체
- 각 픽셀은 색상과 밝기의 미세한 차이를 통해 전체 이미지에 명확함과 섬세함을 부여

## 픽셀

- 작은 사각형의 모자이크
- 그림(picture)과 요소(element)의 합성어
    - 모니터 화면을 자세히 보면 보이는 가장 작은 요소이자 이미지를 확대하면 나오는 아주 작은 단일 점
    
- 단일 픽셀값은 흑백 이미지에서 밝기를 나타내는 숫자
- 컬러 이미지에서 픽셀은 빨강, 녹색, 파랑 등 서로 다른 색상 채널을 나타내는 여러 픽셀 값을 가질 수 있음

## 해상도

- 이미지가 보유하고 있는 픽셀의 양
    - 예) 1920 x 1080 : 이미지가 1920 픽셀의 너비와 1080 픽셀의 높이를 가지고 있음을 의미
    
- 전체 픽셀 수는 이미지의 디테일 양을 결정하는 주요 요소
- 디지털 이미징 초기에는 해상도가 큰 제약 사항이었음

- 이미지 해상도에 영향을 주는 요소들
    - 카메라의 센서 사이즈
        - 디지털 카메라로 촬영한 이미지의 해상도는 주로 카메라의 센서 사이즈에 의해 결정됨
        - 큰 센서는 더 많은 픽셀을 수용할 수 있어, 더 높은 해상도의 이미지를 생성할 수 있음
        
    - 스캐너의 정밀도
        - 물리적 미디어를 디지털화할 때, 해상도는 사용된 스캐너의 정밀도와 품질에 의해 결정됨
        
- 해상도 및 영상 규격을 지칭하는 용어들
    - WVGA : 800 x 480 픽셀로, 과거 동영상 표준 규격
        - 초창기 휴대폰에서 사용하던 규격
        
    - HD : High Definition의 약자. 고화질이라는 의미를 포함
        - 1280 x 720 픽셀을 사용
        - 세로 픽셀 개수에 따라 720p로 표현하기도 함
        
    - FHD : Full HD의 약자
        - 1920 x 1080 픽셀을 사용
        - 영상에서는 1080p로 표현하기도 함
        
    - QHD : Quad HD의 약자
        - 2560 x 1440 픽셀
        - 영상에서는 1440p로 표현
        
    - 4K와 UHD : UHD는 디스플레이 브랜드마다 Ultra HD, Ultra High Definition 등 다른 용어로 표기
        - 4K는 가로 픽셀이 4000 이상을 뜻함
        - 일반적으로 4K UHD를 같이 사용

## 픽셀 밀도

- 디스플레이의 픽셀이 얼마나 촘촘하게 배열되어 있는지를 나타내는 척도
- 일반적으로 인치당 픽셀 수(PPI, Pixel Per Inch), 때로는 센티미터당 픽셀 수로 측정됨
- PPI가 높을수록 주어진 공간에 더 많은 픽셀이 채워져 있음
    - 해상도가 높은 이미지를 보더라도 PPI가 낮다면 이미지가 울퉁불퉁하게 보임
    - 반대로 픽셀 밀도가 높을수록 개별 픽셀의 사이즈가 작아져 육안으로 구분하기 어려움
        - 이미지가 더 부드러워지고 택스트와 그래픽이 더 선명해짐
        
- 픽셀 밀도는 디스플레이의 기술의 발전과 함께 계속해서 중요해지고 있음
    - 초기의 컴퓨터 모니터나 텔레비전은 지금의 디스플레이에 비해 상대적으로 낮은 PPI를 가지고 있었음
    - 모바일 기기와 고해상도 모니터의 등장으로 PPI가 크게 증가
        - 이미지와 텍스트의 선명도를 크게 향상시킴

## 서브 픽셀

- 화면의 픽셀을 구성하는 작은 컬러 요소
- 디스플레이는 RGB 서브 픽셀에서 방출되는 빛의 강도를 조작해서 우리가 보는 색상의 스펙트럼을 재현

- 더 밝고 에너지 효율적인 디스플레이를 추구하면서 RGBW(white) 라고 하는 흰색 서브 픽셀이 추가로 통합됨
    - 흰색 서브 픽셀을 사용하면 디스플레이는 Rgb 서브 픽셀을 완전히 활성화 하지 않고도 흰색을 표현할 수 있으므로 에너지를 절약할 수 있음

## 무손실 압축과 손실 압축

- 고화질 이미지를 표준으로 사용하는 현대 시대에서 이미지 압축은 매우 중요
    - 단순한 저장과 전송을 넘어 컴퓨터 비전 모델의 효율성 및 효과와도 밀접한 관련이 있음
    
    - 효율적인 훈련
        - 압축된 이미지는 모델 훈련 시 I/O 오버헤드를 줄여주므로 데이터 세트가 방대한 경우 프로세스 속도가 빨라짐
        
    - 전송 시간 감소
        - 자율 주행이나 원격 수술과 같은 실시간 애플리케이션에서는 시각 데이터 전송 지연 시간이 매우 중요할 수 있음
        - 압축된 이미지는 더 빠른 전송을 보장하며, AI와 결합하면 더 빠른 의사 결정 프로세스를 촉진할 수 있음
        
    - 컴퓨터 비전 모델 성능 향상
        - 압축 이미지로 AI 모델을 훈련하면 더 강력한 모델을 만들 수 있음
        - 압축으로 인한 미세한 왜곡은 일종의 데이터 증강으로 작용하여 약간 품질이 저하된 이미지에서도 모델이 패턴을 인식하도록 학습시킬 수 있음

### 무손실 압축

- 정보의 손실 없이 데이터 사이즈를 줄이는 것
- 압축이 풀리면 데이터는 원본과 동일한 상태로 돌아감

- PNG(Portable Network Graphics)
    - 웹 그래픽과 이미지 무결성이 가장 중요한 모든 애플리케이션에 선호되는 형식
    - PNG의 핵심 속성
        - 비트 심도 : PNG는 1비트(이진 이미지)에서 16비트에 이르는 다양한 비트 심도를 지원
        - 색상 유형 : 그레이 스케일, RGB, 인덱스 컬러, 알파가 있는 그레이 스케일, 알파가 있는 RGB가 포함됨
        - 투명도 : PNG의 대표적인 기능은 알파 채널 투명도를 지원하여 이진 투명/불투명 뿐만 아니라 미묘한 투명도 수준을 설정할 수 있음
    - PNG 이미지는 디플레이트(deflate)알고리즘을 사용
        - LZ(렘펠-지브)77과 허프만 코딩 알고리즘을 사용
        - LZ77로 데이터의 반복 시퀀스를 효과적으로 축소한 뒤, 허프만 코딩이 그 뒤를 이어 빈도가 높은 패턴을 짧은 코드로 변환함으로써 이미지 파일의 압축률을 극대화

### 손실 압축

- 압축 과정에서 원본 데이터가 일부 손실되는 압축
- 원본 정보의 모든 비트를 보존하는 대신 시각의 한계를 활용하여 감각으로 인지하기 어려운 특정 데이터를 제거
    - 이를 통해 무손실 방식에 비해 훨씬 더 높은 압축률을 달성할 수 있음
    - 특히 이미지, 비디오, 오디오와 같은 미디어에서 유용한 압축 방식
    
- JPEG(Foint Photographic Experts Group)
    - 손실 압축의 가장 유명한 확장자
    - 정지 이미지, 특히 사진을 압축하기 위해 설계 되었음
    - 선 그림이나 글자처럼 가장자리가 날카롭고 대비가 있는 이미지에는 아티팩트(artifact)가 생성되기 때문에 적합하지 않음
    - JPEG 형식은 변환, 양자화 및 엔트로피 코딩의 조합을 사용
        - RGB에서 YCbCr 로 변환(서브 샘플링)
            - Y는 휘도(밝기), Cb와 Cr은 색차(색상)을 나타냄
            
        - 양자화 (quantization)
            - 복잡하고 어려운 픽셀의 패턴은 인간의 시각으로 모두 확인하기 어렵기 때문에 이를 단순화
            - 이 단계에세 손실 압축으로 인한 손실이 발생
            - 얼마나 유지하거나 제거할지 선택할 수 있음
                - 많이 제거하면 이미지 파일은 작아지지만 약간 이상하게 보일 수 있음
                - 조금 제거하면 이미지가 더 좋아 보이지만 파일 용량이 커짐
                
        - 엔트로피 코딩
            - 이미지 세부 정보를 단순화한 후애는 허프만 코딩과 같은 방법을 사용하여 나머지 세부 정보를 효율적으로 저장

# 색 공간

- 픽셀은 이미지의 가장 작은 구성 요소로, 밝기와 색에 대한 정보를 가지고 있음

## 그레이 스케일

- 복잡한 스펙트럼 분포보다는 빛의 강도에 초점을 맞춤
- 색조를 포착하지는 못하지만, 색상을 구별하는 것보다는 명암과 구조적 세부 사항을 강조하는 응용 분야에서 기본적인 표현방법으로 사용됨

- 각 픽셀에 색상 스펙트럼이 없고 대신 밝기를 나타내는 값이 주어짐
    - 0은 빛이 없는 상태(검은색)
    - 255는 최대 밝기(흰색)
    - 단순성 덕분에 많은 계산 시나리오에서 저장 공간을 줄이고 처리 속도를 높일 수 있음
- 가장자리 감지나 텍스처 분석과 같은 특정 이미지 처리 작업의 경우 색상이 방해가 될 수 있으므로 그레이 스케일이 더 적합할 수 있음

## RGB

- 빛의 세 가지 기본 색상
    - 빨강(Red)
    - 초록(Green)
    - 파랑(Blue)
    
- 일상에서 눈으로 보는 거의 모든 색을 표현할 수 있는 기본
- 디지털 이미지에서 세 가지 색상은 채널 이라는 이름으로 표현됨
    - 채널은 각 색상의 정보를 독립적으로 담고 있는 정보의 통로, 또는 저장 공간

## CMYK

- 전자 화면의 발광과는 달리 인쇄물에는 색 표현에 대한 다른 접근이 필요
- 빛을 빼는 원리로 작동
    - 청록색, 자홍색, 노란색 잉크는 종이에서 반사된 흰빛에서 색을 빼는데 사용
    - 이를 결합하여 다양한 색상을 표현
    - 그러나 잉크의 불완전성으로 인해 이들을 조합해도 완벽한 검은색이 나오지 않는 경우가 많으므로 키(검정) 잉크가 포함됨

## HSV

- 인간의 지각에서 파생된 모델
- 우리가 자연스럽게 색을 묘사하는 방식과 더욱 일치하도록 설계됨
    - 사람들은 색을 떠올릴 때 빨간색의 불 같은 느낌이나 파란색의 차분함 등 특정 음영이나 색조를 떠올리는 경우가 많음
    - HSV 공간은 색상 경험을 세 가지 요소로 나누어 이러한 뉘앙스를 포착
        - Hue(색조) : 색의 유형 또는 품질
            - 빨강, 초록, 파랑과 같은 색의 기본 본질
        - Saturation(채도) : 색조의 강도 또는 선명도
            - 채도가 높은 색상은 풍부하고 충만해 보임
            - 채도가 낮은 색상은 색이 바랜 것처럼 보이거나 흐릿해보임
        
        - Value(값, 밝기) : 색상의 밝기 또는 어두움
            - 값이 높으면 색상이 밝고 생생하게 보임
            - 값이 낮으면 어둡거나 칙칙하게 보임

# 비트

- 데이터의 가장 기본적인 단위
- 0 또는 1 두가지 값 중 하나를 가짐
- 모든 디지털 정보는 그 핵심이 비트의 연속으로 표현됨
- 이진수라고도 부름

## 비트 심도

- 이미지의 각 픽셀에 대한 컬러 또는 그레이 스케일 정보를 표현하는 데 사용되는 비트 수

- 1비트 : 두 가지 가능한 값(0 과 1)을 제공
    - 그레이 스케일 이미지에서 회색 음영이 없는 흑백 이미지
    
- 8비트 : 0(검정) 부터 255(흰색) 까지 256개로 표현 가능한 값
    - 부드러운 그라데이션과 그레이 스케일 스펙트럼을 세밀하게 표현할 수 있음
    
- 16비트 : 0부터 65535까지 표현 가능
    - 더욱 부드러운 색을 표현하여 의료 영상과 같이 미묘한 음영 차이가 중요한 전문 환경에서 사용됨
    
- 24비트(채널당 8비트) : 256 x 256 x 256 = 16,777,216 개의 가능한 값
    - 트루 컬러라고도 하며 애플리케이션에 적합한 방대한 색상을 표현할 수 있음
    
- 48비트(채널당 16비트) : 광범위한 색상 값을 제공하여 디지털 아트 및 고화질 사진 등 정확한 색상 표현이 중요한 전문 애플리케이션에 유용함
